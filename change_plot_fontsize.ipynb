# How to change font size in Julia Plots

To change the font sizes in Julia plots, simply add `fontsize` to the attributes. For example, `xlabelfontsize = 12`. 

I learned the trick from [this answer](https://discourse.julialang.org/t/how-to-change-font-size-in-plots-jl/18401/2) by mantzaris.

Here is an example:

```julia
using Plots
```

```julia
x = 1:10
y = 10:19
plot(x, y, xlabel = "x", ylabel = "y", title = "X and Y", label = "y")
```

Here is how we change the font size in the attributes:

```julia
plot!(xlabelfontsize = 12, ylabelfontsize = 14, titlefontsize = 16, labelfontsize = 11)
```
